```python
import numpy as np
import random
from numpy.random import  randint
```

Função fitness

```python
def fitness(ind,distancias):
  
  f = 0 # fitness


  for i in range(len(ind)-1):
    g = ind[i] #cidade em que se encontra
    h = ind[i+1] #proxima cidade
    f += distancias[g][h]

  return f


def retorna_fitness(populacao,tot_pop,distancias):
  
  fit = []
  
  for i in range(tot_pop):
    f = fitness(populacao[i],distancias)
    fit.append(f)

  return fit
```

Seleção por torneio

```python
def selecao_torneio(fitnesses,tot_pop):
  
  ind1 = -1
  ind2 = -1

  while ind1 == ind2:
    # Torneio 1
    sorteados = random.sample(range(0, tot_pop), 2)
    if fitnesses[sorteados[0]] > fitnesses[sorteados[1]]:
      ind1 = sorteados[1]
    else:
      ind1 = sorteados[0]

    # Torneio 2
    sorteados = random.sample(range(0, tot_pop), 2)
    if fitnesses[sorteados[0]] > fitnesses[sorteados[1]]:
      ind2 = sorteados[1]
    else:
      ind2 = sorteados[0]  
        
  return ind1,ind2  
```

Cruzamento de 1 ponto

```python
def arrumaArray(a,n):

  lista = np.arange(1,n) #lista de cidades
  
  b = []
  for i in range(0,n):
    b.append(list())
  
  #Analise da array feita
  for i in lista:
    b[i], = np.where(a == i)
  
  nros = []

  #Quais numeros não estão contidos na array
  for i in range(1,n):
    if(len(b[i]) == 0):
      nros.append(i)

  #Inicializando lista
  repetidos = []
  for i in range(0,n):
    repetidos.append(0)

  #Agora hora de analisar a repeticao(adicionar quantas vezes repete)
  for i in range(1,n):
    l = len(b[i])
    if(l > 1):
      repetidos[i] = l

  i=0
  #Agora é hora de trocar os valores repetidos, por valores que ainda não estão na array
  while i < len(nros):
    cont = 0
    while repetidos[cont] == 0:
      cont += 1
    for j in range(1,repetidos[cont]):
      indice = b[cont][j]
      a[indice] = nros[i]

      i += 1
    repetidos[cont] = 0

  return a

```

```python
def cruzamento(ids,populacao,n):
  
  pai1 = populacao[ids[0]]
  pai2 = populacao[ids[1]]

  ponto = random.randint(1, n-1)

  filho1 = np.concatenate([pai1[:ponto],pai2[ponto:]])
  filho2 = np.concatenate([pai2[:ponto],pai1[ponto:]])

  #É necessário ajustar as arrays, visto que pode haver cidades repetidas

  #Ajusta filho1
  filho1 = arrumaArray(filho1,n)

  #Ajusta filho2
  filho2 = arrumaArray(filho2,n)
  

  return filho1,filho2
```

Elitismo

```python
def elitismo(fitnesses):

  id1 = fitnesses.index(min(fitnesses))

  fitnesses.pop(id1)
  id2 = fitnesses.index(min(fitnesses))

  return id1,id2
```

Mutação

```python
def mutacao(filhos, taxa,n):

  for i in range(len(filhos)):
    if random.random() < taxa:
      pos = random.randint(1, n-1)
      p = random.randint(1,n-1)
      #Troca de posição
      pos2 = np.where(filhos[i] == p)
      filhos[i][pos2] = filhos[i][pos]
      filhos[i][pos] = p

  return filhos
```

```python
n=6
```

```python
#Inicializa a distancia
dist = np.random.randint(1,60, size=(n, n))
dist = np.tril(dist,-1)
for i in range(n):
  for j in range(n):
    dist[i][j] = dist[j][i]
```

```python
#População inicial
num_individuos = 100
populacao_inicial = np.zeros((num_individuos,n+1), dtype=int)

for i in range(num_individuos):
  for j in range(1,n):
    p = random.randint(1,n-1)
    if(p not in populacao_inicial[i]):
      populacao_inicial[i][j] = p
    else:
      p = random.randint(1,n-1)
      while(p in populacao_inicial[i]):
        p = random.randint(1,n-1)
      populacao_inicial[i][j] = p
```
